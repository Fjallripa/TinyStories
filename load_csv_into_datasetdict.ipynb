```python
import datasets
import regex
import numpy as np
from tqdm.auto import tqdm
```

```python
basedir = "/content/drive/MyDrive/genAI project/TinyStoriesTranslate"
files_to_load = {}
r = regex.compile('TS_(.+)_de_(\d+)_(\d+)\.csv')
for filename in os.listdir(basedir):
    split, start, end = r.match(filename).groups()
    files_to_load.setdefault(split, {})[start] = filename

translated_datasets = datasets.DatasetDict({})
for split in files_to_load.keys():
    loading_list = [int(s) for s in files_to_load[split].keys()]
    loading_list.sort()
    dataset_dict = {"text": []}
    progress_bar = tqdm(range(len(loading_list)))
    for start in loading_list:
        #print(f"importing {split} {start}: {files_to_load[split][str(start)]}")
        with open(f"{basedir}/{files_to_load[split][str(start)]}", 'r') as file:
            for story in file:
                dataset_dict["text"].append(story)
        progress_bar.update(1)
    translated_datasets[split] = datasets.Dataset.from_dict(dataset_dict)
print(translated_datasets)
```

```python
translated_datasets.save_to_disk(f"/content/drive/MyDrive/genAI project/translated_dataset_dict")
```
